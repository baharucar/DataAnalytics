```python
import pandas as pd

dataset = pd.read_csv('https://raw.githubusercontent.com/mwaskom/seaborn-data/master/mpg.csv')
dataset
```

Yukarıda verilen veri setini kullanarak görev adımlarında sizden istenen görselleri uygun formatta üretiniz.

>**NOT:** Görsellerinizi istediğiniz kütüphaneyi kullanarak üretebilirsiniz, ayrıca **gerekli tüm görsel ögelerinin görselinizde olmasına** dikkat edin!

## Görev 1:

Veri setinde bulunan arabaları göz önünde bulundurarak, **motor hacimlerinin(displacement) nasıl dağıldığını** uygun görsel yöntemi kullanarak gösterin.

```python
import matplotlib.pyplot as plt
plt.hist(dataset['displacement'], bins=20, edgecolor='black')
plt.title('Motor Hacimlerinin Dağılımı')
plt.xlabel('Motor Hacmi (displacement)')
plt.ylabel('Frekans')
plt.show()
```

## Görev 2:

Veri setinde bulunan arabaları göz önünde bulundurarak, arabaların **menşeilerine(origin) göre ortalama ağırlıklarını(weight) büyükten küçüğe sıralanacak şekilde** uygun görsel yöntemi kullanarak kıyaslayınız.

```python
ortalama_agirlik_orign = dataset.groupby('origin')['weight'].mean().sort_values(ascending=False)
plt.bar(ortalama_agirlik_orign.index, ortalama_agirlik_orign.values)
plt.title('Menşeilerine Göre Ortalama Ağırlık')
plt.xlabel('Menşei')
plt.ylabel('Ortalama Ağırlık')
plt.show()
```

## Görev 3:

Veri setinde bulunan arabaları göz önünde bulundurarak, **arabaların yakıt tüketimi(mpg)** ile **ivmesi(acceleration) arasındaki ilişkiyi** uygun görsel yöntemi kullanarak inceleyin.

```python
import seaborn as sns
sns.scatterplot(x="mpg", y="acceleration", data=dataset)
plt.title('Yakıt Tüketimi ve İvme İlişkisi')
plt.xlabel('Yakıt Tüketimi (mpg)')
plt.ylabel('İvme')
plt.show()
```

## Görev 4:

Veri setinde bulunan arabaları göz önünde bulundurarak, **Ford** ve **Chevrolet** marka araçların **model yıllarına göre ortalama beygir güçlerinin değişimini** uygun görsel yöntemi kullanarak gösterin.

```python
ford_chevrolet = dataset[dataset['name'].str.contains('ford|chevrolet')]
ortalama_degisim = ford_chevrolet.groupby(['model_year', 'name'])['horsepower'].mean().reset_index()

sns.lineplot(x="model_year", y="horsepower", hue="name", data=ortalama_degisim)
plt.title('Ford ve Chevrolet Araçların Model Yılına Göre Ortalama Beygir Gücü')
plt.xlabel('Model Yılı')
plt.ylabel('Ortalama Beygir Gücü')
plt.legend(title='Marka')
plt.show()
```
